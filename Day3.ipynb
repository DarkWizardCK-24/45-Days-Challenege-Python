**range(start, stop, step)**

```python
for i in range(5):            # 0 1 2 3 4
  print(i)
```

```python
for i in range(1, 6):         # 1 to 5
  print(i)
```

```python
for i in range(0, 10, 2):     # 0 2 4 6 8
  print(i)
```

```python
for i in range(5, 0, -1):     # 5 4 3 2 1
  print(i)
```

**Looping Through Lists**

```python
fruits = ["apple", "banana", "mango"]
for item in fruits:
    print(item)
```

**Looping Through Strings**

```python
for ch in "Python":
    print(ch)
```

**break Statement**
* break → Stops the loop immediately.

```python
for i in range(10):
    if i == 5:
        break
    print(i)
```

**continue Statement**
* continue → Skips current iteration.

```python
for i in range(10):
    if i == 5:
        continue
    print(i)
```

**Nested Loops**

```python
for i in range(3):
    for j in range(2):
        print(i, j)
```

```python
#Print sum of even numbers from 1 to 100
total = 0

for i in range(1, 101):
    if i % 2 == 0:
        total += i

print("Sum of even numbers from 1 to 100 is:", total)
```

```python
#Print sum of odd numbers from 1 to 100
total = 0

for i in range(1, 101):
    if i % 2 == 1:
        total += i

print("Sum of odd numbers from 1 to 100 is:", total)
```

```python
#Reverse a String Using a For Loop
text = input("Enter a string: ")

reversed_text = ""

for ch in text:
    reversed_text = ch + reversed_text   # Add character to front

print("Reversed string:", reversed_text)

```

```python
#Print all prime numbers between 1 and 100
for num in range(2, 101):

    is_prime = True

    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break

    if is_prime:
        print(num)
```

```python
#Check if a number is prime
num = int(input("Enter a number: "))

if num <= 1:
    print("Not a prime number")
else:
    is_prime = True

    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break

    if is_prime:
        print("Prime Number")
    else:
        print("Not a Prime Number")
```

```python
#Print Fibonacci series (n terms)
n = int(input("How many terms? "))

a, b = 0, 1

for i in range(n):
    print(a)
    a, b = b, a + b
```

```python
#Print squares of numbers from 1 to n
n = int(input("Enter number n :"))
for i in range(1,n+1):
  q = i*i
  print(q)
```

```python
# Take input from the user
text = input("Enter a string: ")

# Define all vowels (both small and capital)
vowels = "aeiouAEIOU"

# Counter for vowels
count = 0

# Loop through each character in the string
for ch in text:
    # Check if the character is in the vowels list
    if ch in vowels:
        count += 1   # increase the count if vowel found

# Print total number of vowels
print("Number of vowels:", count)
```
