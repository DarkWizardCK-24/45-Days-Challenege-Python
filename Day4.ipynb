**Break Statement**

break â†’ stops the loop immediately

```python
i = 1
while i <= 10:
    if i == 5:
        break
    print(i)
    i += 1
```

**Continue Statement**

continue â†’ skips the current iteration

```python
i = 1
while i <= 5:
    if i == 3:
        i += 1
        continue
    print(i)
    i += 1
```

**Nested While Loops**

```python
i = 1
while i <= 3:
    j = 1
    while j <= 2:
        print(i, j)
        j += 1
    i += 1
```

```python
for i in range(1, 4):
    for j in range(1, 3):
        print(i, j)
```

**Print a multiplication table using while loop**

```python
num = int(input("Enter a number: "))
i = 1

while i <= 10:
    print(num, "x", i, "=", num * i)
    i += 1
```

**Print the sum of first 20 natural numbers**

```python
i = 1
sum = 0

while i <= 20:
    sum += i
    i += 1

print(f"The sum of the first 20 natural numbers is: {sum}")
```

**Count digits in a number using while loop**

```python
num = int(input("Enter number : "))
count = 0

# Use the absolute value to handle potential negative inputs
temp_num = abs(num)

if temp_num == 0:
    count = 1
else:
    # Loop continues as long as the number has digits left
    while temp_num > 0:
        # Removes the last digit
        temp_num = temp_num // 10

        # Counts the digit removed
        count += 1

print(f"The number {num} has {count} digits.")
```

```python
num = int(input("Enter number : "))
reversed_num = 0
temp_num = num # Use a temp variable to preserve the original

while temp_num > 0:
    # 1. Get the last digit (e.g., 54321 % 10 = 1)
    last_digit = temp_num % 10

    # 2. Shift the existing reversed number left, then add the new digit
    reversed_num = (reversed_num * 10) + last_digit

    # 3. Remove the last digit from the temp number (e.g., 54321 // 10 = 5432)
    temp_num = temp_num // 10

print(f"Original: {num}")
print(f"Reversed: {reversed_num}")
```
